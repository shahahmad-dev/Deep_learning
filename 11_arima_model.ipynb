```python
# Import libraries
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Load data
df = pd.read_csv('AirPassengers.csv')

# Call the model
model = ARIMA(df['#Passengers'], order=(2,1,2))

# Fit the model
model_fit = model.fit()

# Make a predication
predication = model_fit.forecast(steps=10)

# Plot the predication
predication.plot()
```
