```python
df = pd.read_csv('AirPassengers.csv')
df.head(10)
```

# **Data Pre-processing**

```python
df = df.rename(columns={
    'Month':'Date',
    '#Passengers': 'Number of Passengers'

})
df
```

```python
# Plot The Data Using Seaborn
fig = plt.figure(figsize=(15, 5))
plt.title('Time Series Analysis', color='Red', fontsize=18,  fontstyle='oblique')

plt.xticks(rotation=60, fontsize=5, color='Blue')
plt.yticks(rotation=0, fontsize=10, color='Green')
sns.lineplot(x = 'Date', y='Number of Passengers',data=df, color='red')
```

```python
# Same Plot Using Plotly
import plotly.express as px
fig = px.line(df, x = 'Date', y='Number of Passengers', title = 'Number of Passanger Over Time')
fig.show()
```
