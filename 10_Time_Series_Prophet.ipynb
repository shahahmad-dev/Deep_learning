```python
import plotly.express as px

fig = px.line(
    df_train_scale,
    x=df_train_scale.index,
    y='meanpressure',
    title='Mean Pressure Over Time'
)

# Format the x-axis to show full date
fig.update_xaxes(
    dtick="D1",              # daily ticks
    tickformat="%d-%m-%Y",   # show day-month-year
    tickangle=45              # tilt labels so they donâ€™t overlap
)

fig.update_layout(
    xaxis_title="Date",
    yaxis_title="Mean Pressure",
    template="plotly_white"
)

fig.show()
```
